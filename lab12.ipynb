```python
#data analysis

#zoe gallagher, csc 201, dataframes, november 29 2022

#dataframes, writing a python module cleans the current data collected 
#read_as_dataframe: This function converts data in a CSV file to a Pandas DataFrame.
#compute_mean: This function computes the mean of a given column in the Pandas DataFrame formatted to two decimal places
#compute_median: This function computes the median of a given column in the Pandas DataFrame formatted to two decimal places.
#get_max: This function computes the maximum of a given column in the Pandas DataFrame formatted to two decimal places.
#get_min: This function computes the minimum of a given column in the Pandas DataFrame formatted to two decimal places.
#output_country_stats: This function outputs the mean, median, maximum and minimum of a given country in the filtered subset of the Pandas DataFrame.
#main: This function setups the program and manages calls to functions defined for specified processing, calculations and output. 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

def read_as_dataframe(filename): 
    temp = pd.read_csv(filename)
    return (temp)

def visually_inspect(df):
    print(df.info())
    
    print('\n\n')
    
    print(df.head())
    
    print('\n\n')
    
    print(df.tail())
    
    print('\n\n')
    
    print(df.columns)
    
    print('\n\n')

def clean_dataframe(df):
    df.drop("Unnamed: 19",axis=1,inplace = True)
    
    df = df.dropna()
    df.drop_duplicates(inplace = True)
    
    df["dem_votes"] = df["dem_votes"].str.replace(',','')
    df['dem_votes'] = pd.to_numeric(df["dem_votes"])
    
    df["rep_votes"] = df["rep_votes"].str.replace(',','')
    df["rep_votes"] = pd.to_numeric(df["rep_votes"])
    
    df["vote_change"] = df["vote_change"].str.replace('%','')
    df["vote_change"] = pd.to_numeric(df["vote_change"])
        
    df.reset_index(drop = True,inplace = True)
    return df

def votes_pivot(subset):
    subset = subset[subset["vote_change"]>=20]
    pivot = subset.groupby("vote_change", as_index = False).median()
    pivot.reset_index(inplace = True)
    return pivot

def main():
    df = read_as_dataframe("popular_vote.csv")
    visually_inspect(df)
    df = clean_dataframe(df)
    visually_inspect(df)
    
    
    x_coords=df["state"]
    y_coords= df["dem_votes"]

    plt.scatter(x_coords,y_coords)
    plt.title("democratic party votes")

    plt.show()
    
    votes_pivot(df)
main()
```
